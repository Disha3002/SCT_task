```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/titanic.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna(subset=["embarked"], inplace=True)
data["deck"].fillna("Unknown", inplace=True)
data["age"].fillna(data["age"].mean(), inplace=True)
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
plt.figure(figsize=(6, 3))
sns.histplot(data["age"], kde=True)
plt.title("Age Distribution")
plt.xlabel("age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.countplot(data=data, x="deck", hue="embark_town")
plt.title("embark_town by deck")
plt.xlabel("deck")
plt.ylabel("embark_town")
plt.legend(title="Survived", loc="upper right")
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.scatterplot(data=data, x="age", y="deck", hue="embark_town")
plt.title("embark_town Plot of age vs deck")
plt.xlabel("age")
plt.ylabel("deck")
plt.legend(title="embark_town")
plt.show()
```
